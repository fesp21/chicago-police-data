```python
import pandas as pd
import numpy as np
from assign_unique_ids_functions import *
```

```python
df = pd.read_csv("../input/salary.csv.gz")
df.head()
```

```python
conflict_cols = ["Start.Date", "Org.Hire.Date", "SPP.Date"]
for year in df['Year'].unique():
    id_cols = ["First.Name", "Last.Name", "Middle.Initial", "Suffix.Name",
               "Gender", "Age.at.Hire","Salary"]
    sub_df = df[df['Year']==year][id_cols + conflict_cols].drop_duplicates()
    sub_df = assign_unique_ids(sub_df, 'TID',id_cols)
    kd = keep_duplicates(sub_df, 'TID')
    print(kd)
    kdg = kd.groupby('TID')[conflict_cols]
    for k,g in kdg:
        #print(g[conflict_cols])
        conflict_list = pd.Series(g[conflict_cols].values.T.ravel())
        conflict_list = conflict_list[conflict_list.notnull()].unique()
        if len(conflict_list) > 1:
            print(conflict_list)
    print(year)
    
```
