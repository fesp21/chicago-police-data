```python
import pandas as pd
from assign_unique_ids_functions import *
from merge_functions import *
pd.options.display.max_columns = 999
```

```python
prof = pd.read_csv('../input/officer-profiles.csv.gz')
pb = pd.read_csv('../input/police-board.csv.gz')
acc = pd.read_csv('../input/accused.csv.gz')
```

```python
acc = acc[(acc['CRID'].isin(pb['CRID'])) | (acc['CRID'].isin(pb['CRID2']))]
```

```python
acc = acc.merge(prof[['UID', 'First.Name', 'Last.Name', 'Middle.Initial', 'Suffix.Name']], on='UID', how='left')
```

```python
keep_duplicates(acc, ['CRID', 'UID'])
```

```python
acc.reset_index(drop=True, inplace=True)
acc.insert(0, 'ind', acc.index + 1)
acc.insert(0, 'CRID2', acc['CRID'])
```

```python
md = merge_datasets(acc, pb, keep_columns = ['ind', 'Case.NO'], extend_base_lists=[['CRID', 'CRID2']],
                   no_match_cols=['Last.Name', 'First.Name'], min_match_length=3,
                   expand_stars=False, F2=False, return_unmatched=True,
                   return_merge_report=True, print_merging=False)
```
