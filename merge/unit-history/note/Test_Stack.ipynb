```python
def number_groups(df, group_id):
    df_nums = []
    current_id = 0
    num = 1
    for i in df[group_id]:
        if i != current_id:
            num = 1
            current_id = i
        else:
            num +=1
        df_nums.append(num)
    df.insert(0, 'Number', df_nums)
    return df
```

```python
uh = number_groups(uh, 'UID')
```

```python
uh.sort_values(['UID', 'Date'], na_position='last',ascending=True, inplace=True)
```
