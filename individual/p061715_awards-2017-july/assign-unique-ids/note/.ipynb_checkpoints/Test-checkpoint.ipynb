```python
def resolve_conflicts(group_df, compare_cols):
    out_df = pd.DataFrame()
    for key, group in group_df:
        conflicts = 0
        total_non_nan = 0
        for col in compare_cols:
            non_nan = group[col].dropna()
            non_nan = non_nan[non_nan != -999]
            total_non_nan += len(non_nan.unique())
            if len(non_nan.unique()) > 1:
                conflicts +=1
        if conflicts == 0 and total_non_nan > 0:
            group.insert(0, 'NID', group.index[0])
            # Not Different
        else:
            group.insert(0, 'NID', group.index)
            # Actually different
        out_df = out_df.append(group)
    return out_df
```
